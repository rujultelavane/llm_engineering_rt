# Tokenizing with tiktoken

```python
import tiktoken

encoding = tiktoken.encoding_for_model("gpt-4.1-mini")

tokens = encoding.encode("Hi my name is Rujul and I like running")

tokens #turn a sentence into tokens
```

```python
for token_id in tokens:
    token_text = encoding.decode([token_id])
    print(f"{token_id} = {token_text}")
    #turn the numbers back into the fragment it represents
```

# Illusion of "memory"

```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

openai = OpenAI()


messages = [
    {"role":"system", "content":"You are a helpful assistant"},
    {"role":"user", "content":"Hi! I'm Rujul!"}
    ]


response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
response.choices[0].message.content
```

```python
messages = [
    {"role":"system", "content":"You are a helpful assistant"},
    {"role":"user", "content":"What's my name?"}
    ]
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
response.choices[0].message.content
```

Every time you call the LLM, it's STATELESS! it has no background from previous calls, only this input sequence. we have to trick it into having memory.

```python
messages = [
    {"role": "system", "content": "You are a helpful assistant"},
    {"role": "user", "content": "Hi! I'm Rujul!"},
    {"role": "assistant", "content": "Hello Rujul! How can I assist you today?"}, #its response back,
    {"role": "user", "content": "What's my name?"}
    ]
response = openai.chat.completions.create(model="gpt-4.1-mini", messages=messages)
response.choices[0].message.content
```

# Recap
1. every LLM call is stateless
2. we pass in the entire convo so far in the input prompt every time
3. this gives the memory illusion
4. but its just a trick - in reality we're giving the entire convo
5. an LLM just predicts the most likely next tokens in the sequence
