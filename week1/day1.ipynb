```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# Load environment variables in a file called .env
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key
if not api_key:
    print("No API key was found")
else:
    print("API key found and looks good so far!")
```

```python
message = "Hello, GPT! This is my first ever message to you! Hi!"

messages = [{"role": "user", "content": message}] ##common dictionary format for messages

messages
```

```python
#calling openai api
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages=messages) ##also a common format (chat.completions.create)
response.choices[0].message.content #prints the response
```

# Notes
models like GPT are trained to recieve instructions a certain way

**system prompt** -- tells them the task and tone to use<br>
**user prompt** -- basically what the user would type (coversation starter to reply to)

# Practice Problem

The code below calls the OpenAI API to produce a subject line for an email, with the user giving the contents of the email.

```python
system_prompt = "Think of a subject line for an email based on the content of the email"
user_prompt = """
    Dear Professor,
    I hope you are having a great week. I am currently working on the final steps of the Midterm Project for our CS114 class, and I had a quick question regarding the data structures requirement mentioned in the rubric.
    Are we required to implement the custom stack-based logic using a dynamic array, or is utilizing the standard built-in list library acceptable as long as we maintain the proper time complexities ($O(1)$ for push and pop operations)? I want to make sure I structure my codebase correctly before submitting.
    I looked through the syllabus and the project discussion board but couldn't quite find a definitive answer. If it is easier to discuss in person, I can also drop by your office hours tomorrow afternoon.
    
    Thank you for your time and help!
    
    Best regards,
    Rujul
"""

messages = [{"role": "system", "content": system_prompt}, {"role": "user", "content": user_prompt}]

#call OpenAI
response = openai.chat.completions.create(model="gpt-4o-mini", messages=messages)

#print the result
print(response.choices[0].message.content)
```
