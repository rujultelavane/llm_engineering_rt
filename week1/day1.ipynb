```python
import os
from dotenv import load_dotenv
from IPython.display import Markdown, display
from openai import OpenAI

# Load environment variables in a file called .env
load_dotenv(override=True)
api_key = os.getenv('OPENAI_API_KEY')

# Check the key
if not api_key:
    print("No API key was found")
else:
    print("API key found and looks good so far!")
```

```python
message = "Hello, GPT! This is my first ever message to you! Hi!"

messages = [{"role": "user", "content": message}] ##common dictionary format for messages

messages
```

```python
#calling openai api
openai = OpenAI()

response = openai.chat.completions.create(model="gpt-5-nano", messages=messages) ##also a common format (chat.completions.create)
response.choices[0].message.content #prints the response
```

# Notes
models like GPT are trained to recieve instructions a certain way

**system prompt** -- tells them the task and tone to use
**user prompt** -- basically what the user would type (coversation starter to reply to)
